```python
import pandas as pd
import numpy as np
```

```python
#Reading the csv files and loading the data-
product = pd.read_csv('/content/Product Data.csv')
sales_exec = pd.read_csv('/content/Sales Executive.csv')
sales = pd.read_csv('/content/Sales.csv')
```

```python
'''
Combine the Data:

Merge the datasets to get a consolidated view of the sales,
with extra columns Product_Name, Category, Name (Sales Executive), and City added to each transaction.

'''
```

```python
#step-1
#merging sales with sales_exec
sales_with_exec = pd.merge(sales, sales_exec, on = 'Sales_Executive_ID', how='left')
```

```python
sales_with_exec
```

```python
#Merging the result obtained from sales_with_exec to product table

final_data = pd.merge(sales_with_exec,product, on = 'Product_ID', how = 'left' )

#final data after merging both the tables
```

```python
final_data
```

```python
'''
Sales Performance Analysis:

Calculate the total sales value for each transaction using the formula:
Sales Value = Quantity * Unit_Price.
Using groupby, determine the top sales executives with the highest total sales value.

'''
#Creating a new column Sales Value by multiplying Quantity and Unit_Price
final_data['Sales Value'] = final_data['Quantity']*final_data['Unit_Price']
```

```python
#final_data.drop(columns = ['Total sales'], inplace=True)
```

```python
final_data
#Final_data after adding a new column named 'Sales Value'
```

```python
fd_grp = final_data.groupby(['Sales_Executive_ID','Name'])['Sales Value']. sum().reset_index()
```

```python
fd_grp
```

```python
fd_grp.sort_values('Sales Value', ascending=False).iloc[0:1]
#Bob is the top sales executives with the highest total sales value.
```

```python
'''
City-Level Insights:

Group the data by City and calculate:
Total sales value per city.
Number of transactions per city.

'''
```

```python
city_grp_by_sales = final_data.groupby('City').agg({'Sales Value':'sum','Transaction_ID':'count'}).reset_index()
city_grp_by_sales
```

```python
'''
Customer Loyalty Analysis:

Identify the top customer (by Customer_ID) who generated the highest total sales.

'''
```

```python
final_data.head()
```

```python
top_cust = final_data.groupby('Customer_ID')['Sales Value'].sum(). reset_index().sort_values('Sales Value', ascending=False).iloc[0:1]
```
